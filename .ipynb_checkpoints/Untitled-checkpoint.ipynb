```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from tensorflow.keras import layers, preprocessing
import pathlib
import tensorflow_hub as hub
from tensorflow.keras.preprocessing import image_dataset_from_directory
import random
import matplotlib.image as mpimg
```

```python
def walk_through_dir(path):
    
    number_of_images = 0
    for dirpath, dirnames, filenames in os.walk(path):
        print(f"There are {len(filenames)} images in '{dirpath}'.")
        number_of_images+=len(filenames)
    print(f"There is {number_of_images} images in our dataset")
```

```python
walk_through_dir("dataset/")
```

```python
def get_classes(path):
    data_dir = pathlib.Path(path)
    class_names = np.array(sorted([item.name for item in data_dir.glob("*")]))
    print(class_names),
    print(f"Number of classes: {len(class_names)}")
    return class_names
```

```python
class_names = get_classes("dataset/")
```

```python
def view_random_images(path, class_names):
    
    
    class_name = random.choice(class_names)
    target_folder = path + class_name
    random_image = random.sample(os.listdir(target_folder), 1)
    image_path = target_folder + "/" + random_image[0]
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Class: {class_name}, shape {img.shape}", fontsize=14)
    
def plot_random_images(path, class_names):
    
    plt.figure(figsize=(25, 25))
    for i in range(16):
        plt.subplot(4,4, i+1)
        img = view_random_images(path, class_names)
```

```python
plot_random_images("dataset/", class_names)
```

```python
data_dir = "dataset/"
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
train_data = image_dataset_from_directory(data_dir,
                                         validation_split=0.2,
                                         batch_size=BATCH_SIZE,
                                         image_size=IMAGE_SIZE,
                                         subset="training",
                                         seed=42)
test_data = image_dataset_from_directory(data_dir,
                                         validation_split=0.2,
                                         batch_size=BATCH_SIZE,
                                         image_size=IMAGE_SIZE,
                                         subset="validation",
                                         seed=42)
```

```python
num_of_classes = {n: 0 for n in range(len(class_names))}
for __,  labels in train_data:
    for label in labels:
        num_of_classes[int(label)] = num_of_classes[int(label)] + 1
```

```python
num_of_classes = pd.DataFrame(num_of_classes.values())
num_of_classes
```

```python
num_of_classes.hist()
```

```python
train_data = train_data.prefetch(buffer_size=tf.data.AUTOTUNE)
train_data
```

```python
test_data = test_data.prefetch(buffer_size=tf.data.AUTOTUNE)
test_data
```

```python
tf.random.set_seed(42)
input_shape = (224, 224, 3)
base_model = tf.keras.applications.EfficientNetB0(include_top=False)
base_model.trainable = False

inputs = layers.Input(shape=input_shape, name='input_layer')
x = base_model(inputs, training = False)
x = layers.GlobalAveragePooling2D(name = 'pooling_layer')(x)
outputs = layers.Dense(len(class_names),activation='softmax', name = 'output_layer')(x)

model_0 = tf.keras.Model(inputs, outputs, name='model_0')
```

```python
model = tf.keras.Sequential([
    # Explicitly define the input shape so the model can be properly
    # loaded by the TFLiteConverter
    tf.keras.layers.InputLayer(input_shape=(224, 224, 3)),
    hub.KerasLayer("https://tfhub.dev/google/efficientnet/b0/feature-vector/1",trainable=False),
    tf.keras.layers.Dense(len(class_names), activation='softmax')
])
model.summary()
```

```python
model_0.compile(loss="sparse_categorical_crossentropy",
              optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              metrics=["accuracy"])
```

```python
model_0.fit(train_data,
         validation_data=test_data,
         epochs = 5)
```
