```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from tensorflow.keras import layers, preprocessing
import pathlib
import tensorflow_hub as hub
from tensorflow.keras.preprocessing import image_dataset_from_directory
import random
import matplotlib.image as mpimg
```

## Data preparations

```python
def walk_through_dir(path):
    
    number_of_images = 0
    for dirpath, dirnames, filenames in os.walk(path):
        print(f"There are {len(filenames)} images in '{dirpath}'.")
        number_of_images+=len(filenames)
    print(f"There is {number_of_images} images in our dataset")
```

```python
walk_through_dir("dataset/")
```

```python
def get_classes(path):
    data_dir = pathlib.Path(path)
    class_names = np.array(sorted([item.name for item in data_dir.glob("*")]))
    print(class_names),
    print(f"Number of classes: {len(class_names)}")
    return class_names
```

```python
class_names = get_classes("dataset/")
```

```python
def view_random_images(path, class_names):
    
    
    class_name = random.choice(class_names)
    target_folder = path + class_name
    random_image = random.sample(os.listdir(target_folder), 1)
    image_path = target_folder + "/" + random_image[0]
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Class: {class_name}, shape {img.shape}", fontsize=14)
    
def plot_random_images(path, class_names):
    
    plt.figure(figsize=(25, 25))
    for i in range(16):
        plt.subplot(4,4, i+1)
        img = view_random_images(path, class_names)
```

```python
plot_random_images("dataset/", class_names)
```

```python
data_dir = "dataset/"
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
```

```python
train_data = image_dataset_from_directory(data_dir,
                                         validation_split=0.2,
                                         batch_size=BATCH_SIZE,
                                         image_size=IMAGE_SIZE,
                                         subset="training",
                                         seed=42)
test_data = image_dataset_from_directory(data_dir,
                                         validation_split=0.2,
                                         batch_size=BATCH_SIZE,
                                         image_size=IMAGE_SIZE,
                                         subset="validation",
                                         seed=42)
```

```python
num_of_classes = {n: 0 for n in range(len(class_names))}
for __,  labels in train_data:
    for label in labels:
        num_of_classes[int(label)] = num_of_classes[int(label)] + 1
```

```python
num_of_classes = pd.DataFrame(num_of_classes.values())
num_of_classes
```

```python
num_of_classes.hist()
```

```python
train_data = train_data.prefetch(buffer_size=tf.data.AUTOTUNE)
train_data
```

```python
test_data = test_data.prefetch(buffer_size=tf.data.AUTOTUNE)
test_data
```

## Callbacks

```python
early_stopping = tf.keras.callbacks.EarlyStopping(patience=5, monitor = "val_loss")
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss",  
                                                 factor=0.2,
                                                 patience=2,
                                                 verbose=1,
                                                 min_lr=1e-7)
```

```python
import datetime
def create_model_checkpoint(dir_name, model_name, save_weights_only = True, monitor = "val_accuracy", save_best_only = True):
    
    path = dir_name + "/" + model_name + ".ckpt"
    model_checkpoint = tf.keras.callbacks.ModelCheckpoint(path,
                                                         monitor=monitor,
                                                         save_best_only=save_best_only,
                                                         save_weights_only=save_weights_only)
    return model_checkpoint
```

```python
def create_tensorboard_callback(dir_name, experiment_name):
    
    log_dir = dir_name + "/" + experiment_name + "/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
    
    tesnorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
    
    print(f"Saving Tensorboard to {log_dir}")
    return tesnorboard_callback
```

## Base Model

```python
import keras_efficientnet_v2
```

```python
input_shape = (224, 224, 3)
base_model = keras_efficientnet_v2.EfficientNetV2B0(num_classes=0, pretrained="imagenet", input_shape = (224, 224, 3))
base_model.trainable = False 


inputs = layers.Input(shape=input_shape, name="input_layer")
preprecess = tf.keras.layers.experimental.preprocessing.Rescaling(1/255.)(inputs)
x = base_model(preprecess)
x = layers.GlobalAveragePooling2D(name="pooling_layer")(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(len(class_names), kernel_initializer="he_normal")(x)
outputs = layers.Activation("softmax", name="softmax_float32")(x) 
model = tf.keras.Model(inputs, outputs, name='EfficientNetB0')
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])
```

```python
history_base_model = model.fit(train_data,
                              validation_data=test_data,
                              epochs=100,
                              callbacks=[early_stopping,
                                        reduce_lr,
                                        create_model_checkpoint("models", "EfficientNetV2B0"),
                                        create_tensorboard_callback("logs", "EfficientNetV2B0")])
```

```python
model_base_performance = model.evaluate(test_data)
model_base_performance
```

## Model with regularization

```python
input_shape = (224, 224, 3)
base_model = keras_efficientnet_v2.EfficientNetV2B0(num_classes=0, pretrained="imagenet", input_shape = (224, 224, 3))
base_model.trainable = False 


inputs = layers.Input(shape=input_shape, name="input_layer")
preprecess = tf.keras.layers.experimental.preprocessing.Rescaling(1/255.)(inputs)
x = base_model(preprecess, training=False)
x = layers.GlobalAveragePooling2D(name="pooling_layer")(x)
x = layers.Flatten()(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(len(class_names), 
                 kernel_initializer="he_normal", 
                 kernel_regularizer=tf.keras.regularizers.l1_l2())(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(len(class_names), 
                 kernel_initializer="he_normal")(x)
outputs = layers.Activation("softmax", name="softmax_float32")(x) 
model_1 = tf.keras.Model(inputs, outputs, name='EfficientNetB0_regularized')
```

```python
model_1.compile(loss="sparse_categorical_crossentropy",
              optimizer=tf.keras.optimizers.Adam(),
              metrics=["accuracy"])
```

```python
history_base_model_regularized = model_1.fit(train_data,
                              validation_data=test_data,
                              epochs=100,
                              callbacks=[early_stopping,
                                        reduce_lr,
                                        create_model_checkpoint("models", "EfficientNetV2B0_regularized"),
                                        create_tensorboard_callback("logs", "EfficientNetV2B0_regularized")])
```
